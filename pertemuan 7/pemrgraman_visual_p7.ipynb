```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik yang diberikan
x1 = 100; y1 = 200
x2 = 800; y2 = 200

x3 = 100; y3 = 900
x4 = 800; y4 = 900

# Lebar garis dan setengah lebar
lw = 10
hw = lw // 2

# Dimensi gambar
row = int(5/7 * 1920)
col = int(5/7 * 1920)

# Buat gambar hitam kosong dengan tiga saluran warna (R,G,B)
Gambar = np.zeros((row, col, 3), dtype=np.uint8)

def titik(x, y):
    # Perbesar titik dengan menggambar lingkaran
    for i in range(x-hw*3, x+hw*3):
        for j in range(y-hw*3, y+hw*3):
            if (i - x)**2 + (j - y)**2 < (hw*3) ** 2:
                Gambar[j, i] = [255, 0, 0]

def garis(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)  # Hitung kemiringan garis

    for x in range(x1-50, x2+50):
        y = int(m * (x - x1) + y1)  # Hitung koordinat y yang sesuai untuk koordinat x menggunakan persamaan garis
        for i in range(x-hw, x+hw+1):
            for j in range(y-hw, y+hw+1):
                Gambar[j, i] = [44, 251, 241]

def vertikal(x1, y1, x2, y2):
    m = (x2 - x1) / (y2 - y1)

    for y in range(y1-50, y2+50):
        x = int(m * (y - y1) + x1)
        for i in range(x-hw, x+hw+1):
            for j in range(y-hw, y+hw+1):
                Gambar[j, i] = [44, 251, 241]

# Horizontal 1
titik(x1, y1)
titik(x2, y2)
garis(x1, y1, x2, y2)

# Horizontla 2
titik(x3, y3)
titik(x4, y4)
garis(x3, y3, x4, y4)

# Vertikal
titik(x1, y1) # Sebenarnya titik itu opsional tapi yasudahlah
titik(x3, y3)
vertikal(x1, y1, x3, y3)

titik(x2, y2)
titik(x4, y4)
vertikal(x2, y2, x4, y4)

# Miring
titik(x1, y1)
titik(x4, y4)
vertikal(x1, y1, x4, y4)


plt.imshow(Gambar)
plt.show()
```

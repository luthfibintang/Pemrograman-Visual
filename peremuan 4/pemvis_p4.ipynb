```python
# Lingkaran
print("\033c");

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-8, 8, 10000)
plt.figure(figsize=(8, 6.5))

y = x -x -0
y1 = (4-x**2)**0.5
y2 = - (4-x**2)**0.5

y3 = 5 + (4-(x-5)**2)**0.5
y4 = 5 - (4-(x-5)**2)**0.5

y5 = -5 + (4-(x-5)**2)**0.5
y6 = -5 - (4-(x-5)**2)**0.5

y7 = 5 + (4-(x+5)**2)**0.5
y8 = 5 - (4-(x+5)**2)**0.5

y9 = -5 + (4-(x+5)**2)**0.5
y10 = -5 - (4-(x+5)**2)**0.5

plt.plot(x, y, '-k')
plt.plot(x, y1, '-r', label='y1, y2')
plt.plot(x, y2, '-r')
plt.plot(x, y3, '-g', label='y3, y4')
plt.plot(x, y4, '-g')
plt.plot(x, y5, '-b', label='y5, y6')
plt.plot(x, y6, '-b')
plt.plot(x, y7, '-y', label='y7, y8')
plt.plot(x, y8, '-y')
plt.plot(x, y9, '-y', label='y9, y10')
plt.plot(x, y10, '-y')

plt.legend(loc = 'upper center')
plt.grid()
plt.show()
```

```python
# Lingkaran
print("\033c");

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-12, 12, 10000)
plt.figure(figsize=(15, 8))

y = x -x -0

y1 = 5 + (25-(x-5)**2)**0.5
y2 = 5 - (25-(x-5)**2)**0.5

y3 = 9 + (4-(x-9)**2)**0.5
y4 = 9 - (4-(x-9)**2)**0.5

y5 = 9 + (4-(x-1)**2)**0.5
y6 = 9 - (4-(x-1)**2)**0.5

plt.plot(x, y, '-k')
plt.plot(x, y1, '-k', label='kepala')
plt.plot(x, y2, '-k')
plt.plot(x, y3, '-k', label='Telinga 1')
plt.plot(x, y4, '-k')
plt.plot(x, y5, '-k', label='Telinga 1')
plt.plot(x, y6, '-k')

plt.fill_between(x, y1, y2, color='black')
plt.fill_between(x, y3, y4, color='black')
plt.fill_between(x, y5, y6, color='black')

plt.legend(loc = 'upper center')
# plt.grid()
plt.show()
```

```python
# Kurva parabolik
print("\033c");

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 10000)
plt.figure(figsize=(3, 9))

plt.plot(x, (0.01-x**2)**0.5, '-k')
plt.plot(x, -((0.01-x**2)**0.5), '-k')

y = x -x -0
y1 = x**3 - 2*x**2 - 5*x + 6

plt.plot(x, y, '-k')
plt.plot(x, y1, '-r', label='y1')

plt.legend(loc = 'upper left')
plt.grid()
plt.show()
```
